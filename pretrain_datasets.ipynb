```python
from pathlib import Path

import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
BASE_DIR = Path("data/")
CXR_LT = BASE_DIR / "cxr-lt-iccv-workshop-cvamd/2.0.0/cxr-lt-2024/train_labeled_no_missing_views.csv"
CXR_LT_26 = BASE_DIR / "cxr-lt-iccv-workshop-cvamd/2.0.0/cxr-lt-2024/development_labeled_task2.csv"
CHEXPERT_TRAIN = BASE_DIR / "chexpert/CheXpert-v1.0/train_cheXbert.csv"
NIH_TRAIN = BASE_DIR / "nih-cxr/Data_Entry_2017.csv"
NIH_TRAIN_LIST = BASE_DIR / "nih-cxr/train_val_list.txt"
VINBIG_TRAIN = BASE_DIR / "vinbig-cxr/train.csv"
```

```python
cxr_lt = pd.read_csv(CXR_LT)
complete_labels = np.array(cxr_lt.iloc[:, -40:].columns)
n_labels = len(complete_labels)
cxr_lt.info()
```

```python
task2_df = pd.read_csv(CXR_LT_26)
complete_labels_26 = np.array(task2_df.iloc[:, -26:].columns)
print(complete_labels_26)
```

```python

other_cols_df = pd.read_csv(CHEXPERT_TRAIN, usecols=range(5))
chexpert_labels_df = pd.read_csv(CHEXPERT_TRAIN, usecols=range(5, 19))
chexpert_labels_df.rename(columns={"No Finding": "Normal"}, inplace=True)

n_chexpert = len(chexpert_labels_df)

print(chexpert_labels_df.info())
chexpert_labels_df.head()
```

```python
chexpert_labels = np.array(chexpert_labels_df.columns)
n_chexpert_labels = len(chexpert_labels)

unmentioned = chexpert_labels_df.isna()
chexpert_labels_df[unmentioned] = -1.0  # Replace NaN (unmentioned) with -1

chexpert_labels_df.head()
```

```python
def add_labels(df: DataFrame, complete_labels: np.ndarray) -> DataFrame:
    """
    Expand the labels in the dataset to match the complete set of labels in cxr-lt.

    """

    labels = np.array(df.columns)

    # discard any existing labels that are not in the complete set
    extras = ~np.isin(labels, complete_labels)
    if extras.any():
        print(labels[extras])
        df = df.drop(columns=labels[extras])
        labels = labels[~extras]

    existing_labels = np.isin(complete_labels, labels)
    non_existing = complete_labels[~existing_labels]

    padding = np.full((len(df), len(non_existing)), fill_value=-1.0)
    new_labels_df = pd.DataFrame(padding, columns=non_existing)

    updated_labels_df = pd.concat([df, new_labels_df], axis=1)
    updated_labels_df = updated_labels_df.reindex(columns=complete_labels)

    return updated_labels_df
```

```python
updated_labels_df = add_labels(chexpert_labels_df, complete_labels)
updated_labels_df_26 = add_labels(chexpert_labels_df, complete_labels_26)
chexpert_df_40 = pd.concat([other_cols_df, updated_labels_df], axis=1)
chexpert_df_26 = pd.concat([other_cols_df, updated_labels_df_26], axis=1)
chexpert_df_40.to_csv(CHEXPERT_TRAIN.parent / "train_expanded.csv", index=False)
chexpert_df_26.to_csv(CHEXPERT_TRAIN.parent / "train_expanded_26.csv", index=False)
print(chexpert_df_40.head().shape)
chexpert_df_26.head()
```

```python
nih_train_val_list = pd.read_table(NIH_TRAIN_LIST, names=["Image Index"])
nih_train_val_list.info()
```

```python
# NIH dataset
nih_df = pd.read_csv(NIH_TRAIN, usecols=range(6))
print(f"Total dataset: {len(nih_df)}")

# get only the train and val set for training
nih_train_idxs = nih_df["Image Index"].isin(set(nih_train_val_list["Image Index"]))
nih_train_df = nih_df[nih_train_idxs].reset_index(drop=True)
print(f"Number of train samples: {len(nih_train_df)}")

nih_labels = nih_train_df.iloc[:, 1].str.split("|").explode().unique()
print(nih_labels)

nih_train_df.head()
```

```python
nih_labels_df = pd.DataFrame(columns=nih_labels, index=nih_train_df.index)
for row, idx in zip(nih_train_df.iloc[:, 1], nih_train_df.index):
    findings = row.split("|")
    for finding in findings:
        nih_labels_df.loc[idx, finding] = 1.0
    nih_labels_df.loc[idx, nih_labels_df.loc[idx].isna()] = 0

nih_labels_df.rename(columns={"No Finding": "Normal"}, inplace=True)
nih_labels_df.rename(columns={"Pleural_Thickening": "Pleural Thickening"}, inplace=True)
nih_labels_df.rename(columns={"Effusion": "Pleural Effusion"}, inplace=True)
nih_labels_df.head()
```

```python
nih_labels_df.sum()
```

```python
updated_labels_df = add_labels(nih_labels_df, complete_labels)
updated_labels_df_26 = add_labels(nih_labels_df, complete_labels_26)
nih_df_40 = pd.concat([nih_train_df, updated_labels_df], axis=1)
nih_df_26 = pd.concat([nih_train_df, updated_labels_df_26], axis=1)
nih_df_40.to_csv(NIH_TRAIN.parent / "train_expanded.csv", index=False)
nih_df_26.to_csv(NIH_TRAIN.parent / "train_expanded_26.csv", index=False)
print(nih_df_40.head().shape)
nih_df_26.head()
```

```python
vinbig_df = pd.read_csv(VINBIG_TRAIN)
vinbig_labels = vinbig_df.iloc[:, 1].unique()
print(vinbig_labels)
vinbig_df.head()
```

```python
vinbig_labels_df = pd.DataFrame(columns=vinbig_labels, index=vinbig_df.index)
for label, idx in zip(vinbig_df.iloc[:, 1], vinbig_df.index):
    vinbig_labels_df.loc[idx, label] = 1.0
    vinbig_labels_df.loc[idx, vinbig_labels_df.loc[idx].isna()] = 0

nodule_mass_idx = vinbig_labels_df.columns.get_loc("Nodule/Mass")
vinbig_labels_df.insert(nodule_mass_idx + 1, "Nodule", vinbig_labels_df["Nodule/Mass"])
vinbig_labels_df.rename(columns={"Nodule/Mass": "Mass"}, inplace=True)
vinbig_labels_df.columns = vinbig_labels_df.columns.str.title()

vinbig_labels_df.rename(columns={"No Finding": "Normal"}, inplace=True)
vinbig_labels_df.rename(columns={"Pulmonary Fibrosis": "Fibrosis"}, inplace=True)

vinbig_labels_df.head()
```

```python
vinbig_df_40 = pd.concat(
    [vinbig_df, add_labels(vinbig_labels_df, complete_labels)],
    axis=1,
)
vinbig_df_26 = pd.concat(
    [vinbig_df, add_labels(vinbig_labels_df, complete_labels_26)],
    axis=1,
)
vinbig_df_40.to_csv(VINBIG_TRAIN.parent / "train_expanded.csv", index=False)
vinbig_df_26.to_csv(VINBIG_TRAIN.parent / "train_expanded_26.csv", index=False)
print(vinbig_df_40.head().shape)
vinbig_df_26.head()
```
